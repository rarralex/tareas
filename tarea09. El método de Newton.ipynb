Recordemos que el *método de Newton* es un método numérico *iterativo* para encontrar raíces de funciones (continuamente) diferenciables.

Sea $f$ la función cuyas raíces queremos encontrar.
La idea es que empecemos desde una adivinanza inicial $x_0$, y que la siguiente aproximación $x_1$ esté donde la recta tangente a la curva $f(x)$ en $x_0$ corte el eje $y$. 

[1] Esboza la geometría, tanto a mano como en la computadora, y así encuentra la expresión de $x_{n+1}$ en términos de $x_n$.

[2] Implementa la iteración para calcular la raíz cuadrada y la raíz cúbica de $2$.  ¿Cuál es una condición razonable de terminación del algoritmo?

[3] Haz un módulo para llevar a cabo diferenciación automática (usando el código del notebook correspondiente) y utilízalo para implementar el método.

Bueno, la ecuación de una recta tangente  una curva, que pasa por el punto $(x_0, f(x_0)$ es $y-f(x_0) = f'(x_0)(x-x_0)$. La siguiente aproximación a la raíz de la función, según el método de Newton, es el cero de la recta tangente, es decir, $x_1 = x_0  - f(x_0)/f'(x_0).$ Si hacemos  lo mismo para $x_1$, es claro que vamos a llegar a una expresión del mismo tipo. De manera que la fórmula para la iteración es $x_{n} = x_{n-1} + f(x_{n-1})/f'(x_{n-1})  $

```julia
using PyPlot
ff(x)=x^2-2
x= EvaluarDerivar[]
a=90.0
N=20
m=3
push!(x,EvaluarDerivar(a,1))
for i in 1:N
    xi= x[i].ev - ff(x[i]).ev/ff(x[i]).der
    push!(x,EvaluarDerivar(xi,1))
end
xx=(0:0.125:100)
ejey=zeros(length(xx))
fx=Float64[]
for i in 1:length(xx)
    push!(fx,ff(xx[i]))
end

y=Array{Float64,1}[]
for j in 1:m
    yj=Float64[]
for i in 1:length(xx)
        p = ff(x[j]).ev + ff(x[j]).der*(xx[i]-x[j].ev)
    push!(yj,p)
end
    push!(y,yj)
end

PyPlot.plot(xx,ejey)
PyPlot.plot(xx,fx, "black")
for i in 1:m
    PyPlot.plot(xx,y[i])
    PyPlot.plot([x[i].ev,x[i].ev],[0,ff(x[i]).ev], linestyle="--")
end
```

```julia
function raiz2(a::Float64)
    f(y)=y^2-a
    n=100
    x=EvaluarDerivar[]
    push!(x,EvaluarDerivar(a,1))
    for i in 2:n
        xi=x[i-1].ev-(f(x[i-1]).ev)/(f(x[i-1]).der)
        push!(x,EvaluarDerivar(xi,1))
    end
    return x[100].ev
end
```

```julia
function raizn(a::Number,n::Int)
    f(y)=y^n-a
    m=100
    x=EvaluarDerivar[]
    push!(x,EvaluarDerivar(a,1))
    for i in 2:m
        xi=x[i-1].ev-(f(x[i-1]).ev)/(f(x[i-1]).der)
        push!(x,EvaluarDerivar(xi,1))
    end
    return x[100].ev
end
```

## Dimensión superior

[5] Desarrolla e implementa el método de Newton para funciones $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^n$. Para hacerlo, toma una adivinaza $\mathbf{x}_n$ y resuelve la ecuación $\mathbf{f}(\mathbf{x}_{n+1}) = \mathbf{0}$, con $\mathbf{x}_{n+1} = \mathbf{x}_n + \delta \mathbf{x}_n$.

Se puede generalizar la fórmula $x_n = x_{n-1} + f(x_{n-1})/f'(x_{n-1})$, simplemente sustituyendo $x_n$ por $\vec{x}_n$, $f(x_{n-1})$ por $\vec{f}(\vec{x}_{n-1})$, y $f'(x_{n-1})$ por $J(\vec{x}_{n-1})$, donde $J$ es la matriz jacobiana de la función, en el punto $x_{n-1}$. Entonces la nueva fórmula es:

$\vec{x}_n = \vec{x}_{n-1} + J^{-1}(\vec{x}_{n-1})*\vec{f}^T(\vec{x}_{n-1})$,

donde $\vec{f}^T$ es el transpuesto de la función, es decir, un vector columna.

[6] Utiliza tu algoritmo para calcular raíces de funciones conocidas multidimensionales.

Lo voy a hacer para la función $f(x,y)=(x^2+y^2-4*x,y^2+2*x-2)$, empezando por el punto (0.5,1), pero no pude hacer una función para calcular la matriz jacobiana de cualquier función. Uso la derivación automatica, pero debo escribir a mano las componentes de la matriz jocobiana para la función dada.

```julia
f(x,y)=[x^2+y^2-4*x,y^2+2*x-2]
f(x::Array{Float64,1})=f(x[1],x[2])
P=Array{Float64,1}[]
push!(P,[0.5,1])
for i in 2:10
       x1=EvaluarDerivar(P[i-1][1],1)
       x2=EvaluarDerivar(P[i-1][2],0)
       y1=EvaluarDerivar(P[i-1][1],0)
       y2=EvaluarDerivar(P[i-1][2],1)
       f1=f(x1,x2)
       f2=f(y1,y2)
       Mjac = inv([f1[1].der f2[1].der ; f1[2].der f2[2].der])
       x=P[i-1]-Mjac*f(P[i-1])
       push!(P,x)
end
P[10]
```
