```julia
function rectangulo(X::Intervalo, Y::Intervalo, c::ASCIIString )
    X1=float64(X.inf)
    X2=float64(X.sup)
    Y1=float64(Y.inf)
    Y2=float64(Y.sup)
    
    plot([X1,X2],[Y1,Y1], color= c)
    plot([X1,X2],[Y2,Y2], color= c)
    plot([X1,X1], [Y1,Y2], color = c)
    plot([X2, X2], [Y1,Y2], color = c)
end
```
