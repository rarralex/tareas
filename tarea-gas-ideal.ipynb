Defino la partícula

```julia
type Particula
    x::Vector{Float64}
    v::Vector{Float64}
    m::Float64
end
```

```julia
srand(123)
```

```julia
function mover!(dt,p::Particula)
p.x+=p.v*dt
end
```

```julia
function crear_particula()
Particula(rand(2),rand(2),1)
end
```

Le puse masa = 1 a todas las partículas, porque por el momento la masa no entra en la dinámica.

NO le puse la masa al gas porque nunca la usé. Además para saber el número de partículas, puedo hacer una función que me diga la longitud del vector particulas, dentro de Gas.

```julia
type Gas
    particulas::Vector{Particula}
end
```

```julia
function NumPartGas(g::Gas)
    length(g.particulas)
end
```

```julia
function Gas(N::Int)
particulas=Particula[]
for i in 1:N
push!(particulas,crear_particula())
end
Gas(particulas)
end
```

```julia
g1=Gas(2)
```

```julia
function vaciar(v)
    l=length(v)
    for i in 1:l
        pop!(v)
    end
end
```

```julia
puntosx1=Float64[]
puntosy1=Float64[]
puntosx2=Float64[]
puntosy2=Float64[]
function GasIcaja(Lx::Float64,Ly::Float64,N::Int,dt::Float64,g::Gas)
    vaciar(puntosx1)
    vaciar(puntosy1)
    vaciar(puntosx2)
    vaciar(puntosy2)
    entradas=g.particulas #es un vector que en cada entrada tiene un tipo Particula
    l=length(entradas)
    for j in 1:N
        mover!(dt,g)
        for i in 1:l  #Voy a reflejar a las partículas sobre la pared
            if (entradas[i].x[1]<=0 || entradas[i].x[1]>=Lx) # si choca con las paredes en x=0,L, vx=-vx
            entradas[i].v=[-entradas[i].v[1],entradas[i].v[2]]
        end
            if (entradas[i].x[2]<=0 || entradas[i].x[2]>=Ly) # lo mismo, si chocha en y, cambia sólo vy.
            entradas[i].v=[entradas[i].v[1],-entradas[i].v[2]]
        end
        end
        push!(puntosx1,entradas[1].x[1]) #aquí se decide que partícula seguir
        push!(puntosy1,entradas[1].x[2]) #aquí se decide que partícula seguir
        push!(puntosx2,entradas[2].x[1]) #aquí se decide que partícula seguir
        push!(puntosy2,entradas[2].x[2]) #aquí se decide que partícula seguir
end
#puntosx1
#puntosy1
#puntosx2
#puntosy2
#Gas1(g.particulas)
end
```
