## 1. Explicación de método presentado por Tucker

Dada una ecuación diferencial con condiciones inciales:

$$\dot{x}(t)=f(x(t),t),   \  \  \  \ \  \  \  \ \  t\in [0,T]  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (1) $$

$$x(0) = x_0.$$
Esta ecuación, tiene solución única en el intervalo $0<x<h$, $0<t<T'$, donde $h <= D$ y $T' <= T$; los intervalos $[0,D]$, $[0,T]$ son aquellos donde $f$ y $\partial f/ \partial t$ son continuas.


Se busca un método para encerrar (con intervalos) la función $x(t)$, solución de la ecuación diferencial planteadad. Para ello hay que considerar a $f$ como un flujo $\varphi(x,t)$, de manera que la ecuación $(1)$ debe reescribirse como $$ \frac{\partial}{\partial t} \varphi(x,t) = f(\varphi(x,t),t)  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (2)$$
$$\varphi(x,0) = x.$$

Particularizando al caso donde $\varphi$ (y $f$) no depende del tiempo, la ecuación $(2)$ puede escribirse en forma integral 
$$\varphi(x_0,t)=x_0 + \int_0^t f(\varphi(x_0,s))ds.$$

Si suponemos que $\Psi(t)$ es una "función de intervalos" continua que encierra la solución $x$, entonces por inclusión isotónica $$x(t) \in x_0 + \int_0^t F(\Psi(s))ds, \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (3)$$ donde $F$ es la extensión a intervalos de $f$.

Podemos construir $\Psi$ a partir de una solución aproximada por otro método (sumando el error dado por el método, hacia arriba y hacia abajo).

```julia
figura2= imread("fig2.png")
```

La cuestión es si $(3)$ puede servir para calcular otro continuio de intervalos que se aproximen más a $f$. Tucker demuestra que el algoritmo que permite acercarse a la solución dado un intervalo inicial $\Psi(0) = z(0)$ es 
$$ z^{(k+1)} = x_0 + \int_0^t F(z^{k})= x_0 + [0,t] \times F(z^{(k)}), \ \ \ \ \ \ \ \ \ \ (4)$$
donde $t \in [0, T']$ y $T' = min\{ T, D/M, 1/2K\}$, según el teorema de existencia y unicidad de las ecuaciones diferenciales; $M$ es una cota para $f$ en el rectangulo $[0,D] \times [0,T]$, y $K$ es "la constante de Lipschitz" para la función $f$. 

Esta aproximación se puede refinar por el cono que se obtiene al variar t en la ecuación
$$ x_0 + t F(z^{(k)})\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (5)$$

## 3. Primer paso

Empecé con el problema $$\dot{x} = x$$ $$ x(0)=1,$$ cuya solución es $$x(t)=exp(x).$$

La función generalizada a intervalos a considerar es:

```julia
F(x) = x

#SOLUCIÓN

f(x) = exp(x)
```

Consideré el intervalo inicial, y la posición inicial

```julia
using intervalos
using FactCheck
include("test-integracion.jl")
Z0= Intervalo(0.5, 1.5)
x0=1
```

La función es Lipschitz, porque  $$|F(x_1)-F(x_2)| = |x_1-x_2| <= K |x_1 - x_2 | $$si y sólo si $$K>=1.$$ Entonces tomo $K=2$. Ahora, la cota para $F(x)$ en el intervalo $Z0$ es $1.5$, luego $$T'= min \{\frac{1.5}{1.5}, \frac{1}{2*2}\} = 1/4.$$
Eso significa que los pasos de tiempo serán de $0.25$, sin tener que variarlo conforme nos movemos.

```julia
dt=0.25
T=Intervalo(0,dt)
Z=Intervalo[]
push!(Z,Z0)
for i in 2:10
    Zi=x0 + T*F(Z[i-1])
    push!(Z,Zi)
end
AA=x0+dt*F(Z[5])
```

#### Gráfica de lo anterior

```julia
a = (0:0.0125:0.3)
b = Float64[]
for i in 1:length(a)
    push!(b,f(a[i]))
end

plot(a,b)
caja(T,Z[5],"red")
plot(inter(T),[x0,float64(AA.sup)], "red")
plot(inter(T),[x0,float64(AA.inf)], "green")
```

## 3.3 Paso n

Para realizar n pasos definí usé el algoritmo

```julia
#=
function npaso1(F::Function, f::Function, Z0::Intervalo, x0::Float64, dt::Float64, N::Int, M::Int, asup::Float64)
    X0=Float64[]
    T=Intervalo[]
    Z=Intervalo[]
    AA=Intervalo[]
    k=-1
    for j in 1:N
        ZZ=Intervalo[]

        if (j ==1)
            push!(X0,x0)
            push!(ZZ,Z0)
            push!(T,Intervalo(0,dt))
            for i in 2:M
                Zi=X0[j] + T[j]*F(ZZ[i-1])
                push!(ZZ,Zi)
            end
            push!(Z,ZZ[M])
            BB=X0[j] + dt*F(Z[j])
            push!(AA,BB)
        end

        if (j==2)
            push!(X0,float64(AA[j-1].sup))
            push!(ZZ,AA[j-1])
            push!(T,Intervalo((j-1)*dt,j*dt))
            for i in 2:M
                Zi=X0[j]+ T[j]*F(ZZ[i-1])
                push!(ZZ,Zi)
            end
            push!(Z,ZZ[M])
            BB=X0[j]+dt*F(Z[j])
            push!(AA,BB)

            ZZ=Intervalo[]
            push!(X0,float64(AA[j-1].inf))
            push!(ZZ,AA[j-1])
            for i in 2:M
                Zi=X0[j+1]+ T[j]*F(ZZ[i-1])
                push!(ZZ,Zi)
            end
            push!(Z,ZZ[M])
            BB=X0[j+1] + dt*F(Z[j+1])
            push!(AA,BB)
        end

        if (j>2)
            k+=1
            push!(X0,float64(AA[j-1+k].sup))
            push!(ZZ, Intervalo(AA[j-1+k].sup,AA[j+k].inf))
            push!(T,Intervalo((j-1)*dt,j*dt))
            for i in 2:M
                Zi= X0[j+1+k] + T[j]*F(ZZ[i-1])
                push!(ZZ,Zi)
            end
            push!(Z,ZZ[M])
            BB=X0[j+1+k]+dt*F(ZZ[M])
            push!(AA,BB)

            ZZ=Intervalo[]
            push!(X0,float64(AA[j+k].inf))
            push!(ZZ, Intervalo(AA[j-1+k].sup,AA[j+k].inf))
            for i in 2:M
                Zi=X0[j+2+k]+T[j]*F(ZZ[i-1])
                push!(ZZ,Zi)
            end
            push!(Z,ZZ[M])
            BB=X0[j+2+k]+dt*F(ZZ[M])
            push!(AA,BB)
        end
    end

    ainf=0
   # asup=float64(T[N].sup)
    a = (ainf:0.0125:asup)
    b = Float64[]
    for i in 1:length(a)
        push!(b,f(a[i]))
    end

    #GRÁFICA FUNCIÓN
    plot (a,b)


    k=-1
    for i in 1:N
        if (i==1)
            plot(inter(T[i]), [X0[i],float64(AA[i].sup)], "red")
            plot(inter(T[i]),[X0[i],float64(AA[i].inf)], "green")
        else
        k+=1
            plot(inter(T[i]), [X0[i+k],float64(AA[i+k].sup)], "red")
            plot(inter(T[i]),[X0[i+1+k],float64(AA[i+1+k].inf)], "green")
        end
    end
    
end # function
=#
```

* npaso1 de:
    * Función $F$ que define la ecuación diferencial
    0. Función solución exacta (sólo para comprobar), $f$
    3. Punto inicial $x_0$
    4. Paso de tiempo $dt$
    5. Número de pasos, $N$
    6. Número de las iteraciones de $Z^{(n)}$, $M$
    7. Punto final de la gráfica de $f$, $asup$
